***5.1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)***

Null Hypothesis: Both average MEDV are the same

Alternative Hypothesis: Both average MEDV are NOT the same

```python
boston_df["CHAS"].value_counts()
```

```python
x = boston_df[boston_df["CHAS"] == 0]["MEDV"]
x
```

```python
y = boston_df[boston_df["CHAS"] == 1]["MEDV"]
y
```

```python
scipy.stats.ttest_ind(x,y,axis=0,equal_var=True)
```

pvalue is > Fvalue (0.005) hence fail to reject the null hypothesis. Thus Alternative Hypothesis: Both average MEDV are not the same

***5.3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)***







State the hypothesis

 Null-hypothesis: NOX is not correlated with INDUS

 Alternative-hypothesis: NOX is correlated with INDUS

```python
pearson,p_value = scipy.stats.pearsonr(boston_df["NOX"],boston_df["INDUS"])
```

```python
print("Pearson Coefficient value={0}, P-value={1}".format(pearson,p_value))
```

Pvalue < 0.05, thus Null-hypothesis is rejected. Thus NOX is correlated with INDUS
